```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#loading the output
df0 = pd.read_csv('./Project/out/data_0.csv') #Z0.02
df1 = pd.read_csv('./Project/out/data_1.csv') #Z0.00014
```

```python
df0_merge_0 = df0[df0['Merge'] == 0]
df0_merge_1 = df0[df0['Merge'] == 1]
df1_merge_0 = df1[df1['Merge'] == 0]
df1_merge_1 = df1[df1['Merge'] == 1]

sns.set(style='whitegrid')

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

ax1.scatter(df0_merge_0['Mass_0'], df0_merge_0['Mass_1'], s=10, c='blue', edgecolor='k', label='Non-Merging systems', alpha=0.7)
ax1.scatter(df0_merge_1['Mass_0'], df0_merge_1['Mass_1'], s=10, c='red', edgecolor='k', label='Merging systems', alpha=0.7)
ax1.set_xlabel('Mass_0', fontsize=14)
ax1.set_ylabel('Mass_1', fontsize=14)
ax1.set_title('Z = 0.02', fontsize=16, fontweight='bold')


ax2.scatter(df1_merge_0['Mass_0'], df1_merge_0['Mass_1'], s=10, c='blue', edgecolor='k', label='Non-Merging systems', alpha=0.7)
ax2.scatter(df1_merge_1['Mass_0'], df1_merge_1['Mass_1'], s=10, c='red', edgecolor='k', label='Merging systems', alpha=0.7)
ax2.set_xlabel('Mass_0', fontsize=14)
ax2.set_ylabel('Mass_1', fontsize=14)
ax2.set_title('Z = 0.00014', fontsize=16, fontweight='bold')

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 6), sharey=False)

axes[0,0].hist(df0_merge_0['Semimajor'], bins=30, color='blue', edgecolor='black', alpha=0.7)
axes[0,0].set_title('Non-Merging systems', fontsize=14)
axes[0,0].set_xlabel('Semimajor', fontsize=12)
axes[0,0].set_ylabel('Frequency', fontsize=12)

axes[0,1].hist(df0_merge_1['Semimajor'], bins=30, color='red', edgecolor='black', alpha=0.7)
axes[0,1].set_title('Merging systems', fontsize=14)
axes[0,1].set_xlabel('Semimajor', fontsize=12)

axes[1,0].hist(df1_merge_0['Semimajor'], bins=30, color='blue', edgecolor='black', alpha=0.7)
axes[1,0].set_title('Non-Merging systems', fontsize=14)
axes[1,0].set_xlabel('Semimajor', fontsize=12)
axes[1,0].set_ylabel('Frequency', fontsize=12)

axes[1,1].hist(df1_merge_1['Semimajor'], bins=30, color='red', edgecolor='black', alpha=0.7)
axes[1,1].set_title('Merging systems', fontsize=14)
axes[1,1].set_xlabel('Semimajor', fontsize=12)

fig.text(0.5, 1, 'Z = 0.02', ha='center', va='center', fontsize=16, fontweight='bold')
fig.text(0.5, 0.5, 'Z = 0.0014', ha='center', va='center', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 6), sharey=False)

count_stable_0 = [df0_merge_0['MTEvents_stable'].sum(), df1_merge_0['MTEvents_stable'].sum()]
count_unstable_0 = [df0_merge_0['MTEvents_unstable'].sum(), df1_merge_0['MTEvents_unstable'].sum()]
count_stable_1 = [df0_merge_1['MTEvents_stable'].sum(), df1_merge_1['MTEvents_stable'].sum()]
count_unstable_1 = [df0_merge_1['MTEvents_unstable'].sum(), df1_merge_1['MTEvents_unstable'].sum()]

x_values = [0, 0.5]
labels = ['Stable', 'Unstable']
heights_0 = [[count_stable_0[0], count_unstable_0[0]],[count_stable_0[1], count_unstable_0[1]]]
heights_1 = [[count_stable_1[0], count_unstable_1[0]],[count_stable_1[1], count_unstable_1[1]]]


axes[0,0].bar(x_values, heights_0[0], width=0.4, color=['blue', 'red'])
axes[0,0].set_title('Merging Systems', fontsize=14)
axes[0,1].bar(x_values, [count_stable_1[0], count_unstable_1[0]], width=0.4, color=['blue', 'red'])
axes[0,1].set_title('Non-Merging Systems', fontsize=14)
axes[0,0].set_xticks(x_values)
axes[0,0].set_xticklabels(labels)
axes[0,1].set_xticks(x_values)
axes[0,1].set_xticklabels(labels)

axes[1,0].bar(x_values, heights_0[1], width=0.4, color=['blue', 'red'])
axes[1,0].set_title('Merging Systems', fontsize=14)
axes[1,1].bar(x_values, [count_stable_1[1], count_unstable_1[1]], width=0.4, color=['blue', 'red'])
axes[1,1].set_title('Non-Merging Systems', fontsize=14)
axes[1,0].set_xticks(x_values)
axes[1,0].set_xticklabels(labels)
axes[1,1].set_xticks(x_values)
axes[1,1].set_xticklabels(labels)


fig.suptitle('Frequency of Stable and Unstable Mass Transfer Events', fontsize=16, fontweight='bold')

fig.text(0.5, 0.91, 'Z = 0.02', ha='center', va='center', fontsize=16, fontweight='bold')
fig.text(0.5, 0.45, 'Z = 0.0014', ha='center', va='center', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```
